```python
import pathlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = pathlib.Path('.').absolute().parent.joinpath('data/cowley_ltn_busgate_fines.csv')
```

```python
df = pd.read_csv(data_path)
```

```python
df
```

```python
ax = df.plot(x="Date range", y=["Bartholomew Road", "Cornwallis Road"], kind="bar", figsize=(10,8))
ax.axvline(x=5, ymin=0, ymax=1800, label="Bartholomew Road fines issued from 01/03/2022",color="C0", linestyle="--")
ax.axvline(x=3, ymin=0, ymax=1800, label="Cornwallis Road fines issued from 14/02/2022",color="C1", linestyle="--")
ax.arrow(12, 700, 0, -200, head_width=0.5, head_length=100, color='black')
ax.text(11.5, 750, "Easter?")
ax.set_title("Bus gate fines in the Cowley LTN")
ax.legend()
ax.figure.savefig('cowley_ltn_bus_gates.png', bbox_inches='tight')
```

- Bartholomew Road:
  - Commenced on 04/02/2022 and initially Warning notices were issued. 
  - Live Penalty Charge Notices were issued from 01/03/2022
- Cornwallis Road:
  - Commenced on 24/01/2022 and initially Warning notices were issued. 
  - Live Penalty Charge Notices were issued from 14/02/2022
